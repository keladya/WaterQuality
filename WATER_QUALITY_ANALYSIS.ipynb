# Analisis Eksplorasi dan Visualisasi

## Comparison (Perbandingan)

```python
# Mean value

df.groupby('Color')[['pH', 'Iron', 'Nitrate', 'Chloride', 'Lead', 'Zinc',
                     'Turbidity', 'Fluoride', 'Copper', 'Odor', 'Sulfate',
                     'Conductivity', 'Chlorine', 'Manganese',
                     'Total Dissolved Solids', 'Water Temperature', 'Air Temperature']].mean().sort_values(by='pH', ascending=False)
```

```python
# Mengelompokkan data berdasarkan Bulan, dan menghitung rata-rata
df_grouped = df_clean.groupby('Month')[['pH', 'Iron', 'Nitrate', 'Chloride', 'Lead',
                                  'Zinc', 'Turbidity', 'Fluoride', 'Copper', 'Odor', 'Sulfate',
                                  'Conductivity', 'Chlorine', 'Manganese',
                                  'Total Dissolved Solids', 'Water Temperature', 'Air Temperature']].mean()

# Mengurutkan dan memilih 6 nilai tertinggi dan terendah berdasarkan 'pH' (misalnya)
top_6 = df_grouped.sort_values(by='pH', ascending=False).head(6)
bottom_6 = df_grouped.sort_values(by='pH', ascending=True).head(6)
bottom_6 = bottom_6.sort_values(by='pH', ascending=False)

# Menggabungkan top 6 dan bottom 6 untuk plotting
highlighted_data = pd.concat([top_6, bottom_6])

# Mendefinisikan warna: merah untuk top 6 (pH tinggi), hijau untuk bottom 6 (pH rendah)
colors = ['red'] * 6 + ['green'] * 6

# Membuat plot batang
plt.figure(figsize=(12, 6))
sns.barplot(x=highlighted_data.index, y=highlighted_data['pH'], palette=colors)

# Menambahkan judul dan label sumbu
plt.title('6 Bulan Teratas dan Terbawah Berdasarkan Nilai Rata-rata pH', fontsize=16)
plt.xlabel('Bulan', fontsize=12)
plt.ylabel('Rata-rata Nilai pH', fontsize=12)
plt.xticks(rotation=45)

# Menampilkan plot
plt.tight_layout()
plt.show()
```

```python
# Memfilter data berdasarkan Bulan
agustus_data = df_clean[df_clean['Month'] == 'August']
februari_data = df_clean[df_clean['Month'] == 'February']

# Memilih kolom yang relevan untuk satu baris data
agustus_values = agustus_data[['pH', 'Iron', 'Nitrate', 'Chloride', 'Lead',
                                  'Zinc', 'Turbidity', 'Fluoride', 'Copper', 'Odor', 'Sulfate',
                                  'Conductivity', 'Chlorine', 'Manganese',
                                  'Total Dissolved Solids', 'Water Temperature', 'Air Temperature']].iloc[0]
februari_values = februari_data[['pH', 'Iron', 'Nitrate', 'Chloride', 'Lead',
                                  'Zinc', 'Turbidity', 'Fluoride', 'Copper', 'Odor', 'Sulfate',
                                  'Conductivity', 'Chlorine', 'Manganese',
                                  'Total Dissolved Solids', 'Water Temperature', 'Air Temperature']].iloc[0]

# Menggabungkan data ke dalam satu DataFrame untuk plotting
comparison_df = pd.DataFrame({
    'Parameter Kualitas': agustus_values.index,
    'Agustus': agustus_values.values,
    'Februari': februari_values.values
})

# Mengubah DataFrame ke format long agar mudah di-plot dengan seaborn
comparison_df_melted = comparison_df.melt(id_vars='Parameter Kualitas', var_name='Bulan', value_name='Nilai Kualitas Air')

# Membuat plot batang
plt.figure(figsize=(12, 8))
sns.barplot(x='Parameter Kualitas', y='Nilai Kualitas Air', hue='Bulan', data=comparison_df_melted, palette={'Agustus': 'red', 'Februari': 'green'})

# Menambahkan judul dan label sumbu
plt.title('Perbandingan Kualitas Air Antara Agustus dan Februari', fontsize=16)
plt.xlabel('Parameter Kualitas Air', fontsize=12)
plt.ylabel('Nilai Kualitas Air', fontsize=12)

# Membuat label sumbu x miring dan memutarnya agar tidak bertumpuk
plt.xticks(rotation=45, fontstyle='italic', ha='right')

# Menampilkan plot
plt.tight_layout()
plt.show()
```

## Composition (Komposisi)

```python
# Mengelompokkan pH menjadi 'Asam' dan 'Basa' menggunakan pd.cut
df_clean['Kategori pH'] = pd.cut(df_clean['pH'], bins=[0, 7, float('inf')], labels=['Asam', 'Basa'], right=False)

# Menghitung jumlah masing-masing kategori
kategori_counts = df_clean['Kategori pH'].value_counts()

# Membuat pie chart berdasarkan kategori pH ('Asam' dan 'Basa')
plt.figure(figsize=(6, 6))
plt.pie(
    kategori_counts,
    labels=kategori_counts.index,
    autopct='%1.1f%%',
    startangle=90,
    colors=['#66b3ff', '#ff9999']  # Warna biru untuk 'Asam' dan merah muda untuk 'Basa'
)

# Menambahkan judul
plt.title('Distribusi Kategori pH (Asam vs Basa)', fontsize=14)

# Mengatur pie chart
plt.axis('equal')

# Menampilkan chart
plt.tight_layout()
plt.show()
```

## Distribution (Distribusi)

```python
# Membuat palet warna khusus yang dibalik (inverted)
original_palette = sns.color_palette('RdYlGn', n_colors=2)  # 2 warna untuk kategori pH
inverse_palette = original_palette[::-1]  # Membalik urutan warna

# Membuat boxplot untuk distribusi pH berdasarkan kategori pH
plt.figure(figsize=(12, 8))
sns.boxplot(x='Kategori pH', y='pH', data=df_clean, palette=inverse_palette)

# Menambahkan judul dan label
plt.title('Boxplot Distribusi Nilai pH Berdasarkan Kategori', fontsize=16)
plt.xlabel('Kategori pH', fontsize=12)
plt.ylabel('Nilai pH', fontsize=12)
plt.xticks(rotation=45)

# Mengatur tata letak dan menampilkan plot
plt.tight_layout()
plt.show()
```

```python
# Membuat histogram untuk distribusi pH dengan warna yang disesuaikan
plt.figure(figsize=(10, 6))
sns.histplot(data=df_clean, x='pH', hue='Kategori pH', bins=30, kde=True, palette=['#ff9999', '#66b3ff'])  # Asam -> Merah Muda, Basa -> Biru

# Menambahkan judul dan label
plt.title('Distribusi Nilai pH pada Sampel', fontsize=16)
plt.xlabel('Nilai pH', fontsize=12)
plt.ylabel('Frekuensi', fontsize=12)

# Menampilkan plot
plt.tight_layout()
plt.show()
```

Berdasarkan grafik histogram yang dihasilkan untuk distribusi nilai pH dalam dua kategori, yaitu Asam (pH < 7) dan Basa (pH > 7). Pada grafik, terlihat bahwa distribusi untuk kategori Asam memiliki puncak di sekitar nilai pH lebih dari 6, sementara kategori Basa memiliki puncak di sekitar nilai pH antara 7 sampai 8. Distribusi nilai pH yang berada dalam kategori Basa lebih mendominasi sampel dengan frekuensi yang lebih tinggi dibandingkan dengan kategori Asam.

Distribusi pH dalam kategori Basa cenderung meluas, dengan sebagian kecil sampel memiliki nilai pH lebih dari 9. Sementara itu, pada kategori Asam, nilai pH lebih terkonsentrasi di rentang yang lebih sempit, dengan sedikit sampel yang memiliki nilai pH di bawah 5. Hal ini menunjukkan bahwa sebagian besar sampel yang diuji memiliki sifat basa, dan hanya sebagian kecil yang memiliki sifat asam. Dengan demikian interpretasi ini sesuai dengan grafik pie chart yang telah dilakukan sebelumnya.

#Relationship (Hubungan Antar Variabel)

```python
# List variabel kecuali pH untuk perbandingan
comparison_variables = ['Water Temperature', 'Conductivity', 'Fluoride', 'Odor', 'Nitrate',
                        'Chloride', 'Zinc', 'Turbidity', 'Copper', 'Iron', 'Chlorine',
                        'Sulfate', 'Manganese', 'Air Temperature']

# Plotting pH vs variabel lainnya
for variable in comparison_variables:
    # Menghitung korelasinya
    correlation = df[['pH', variable]].corr().loc['pH', variable]

    # Plotting plot regresi
    plt.figure(figsize=(10, 6))
    sns.regplot(x='pH', y=variable, data=df)

    # Menambah judul dan label
    plt.title(f'Regression Plot of pH vs {variable}', fontsize=16)
    plt.xlabel('pH', fontsize=12)
    plt.ylabel(variable, fontsize=12)

    # Menampilkan koefisien korelasi pada plot
    plt.text(
        x=df['pH'].min(),  # X koordinat untuk text
        y=df[variable].max() * 0.9,  # Y koordinat untuk text
        s=f'Correlation: {correlation:.2f}',  # Text to display
        fontsize=12,
        color='red'
    )

    # Menunjukkan plot
    plt.tight_layout()
    plt.show()
```

1. pH vs Water Temperature:

Hubungan antara suhu air dan pH sangat lemah. Hal ini menunjukkan bahwa tingkat keasaman atau kebasaan air tidak banyak dipengaruhi oleh suhu.

2. pH vs Conductivity:

Korelasi negatif lemah terlihat antara konduktivitas dan pH, mengindikasikan bahwa pH cenderung sedikit lebih rendah saat konduktivitas tinggi. Hal ini mungkin karena kandungan ion terlarut yang lebih banyak, yang dapat menurunkan pH.

3. pH vs Fluoride:

Tidak ada hubungan yang signifikan antara fluoride dan pH, menunjukkan bahwa kadar fluoride tidak secara langsung mempengaruhi tingkat keasaman air.

4. pH vs Odor:

Tidak terdapat korelasi yang signifikan antara bau dan pH, yang menandakan bahwa karakteristik bau air tidak berkaitan langsung dengan keasaman atau kebasaan air.

5. pH vs Nitrate:

Hubungan yang sangat lemah terlihat antara nitrat dan pH. Meskipun nitrat seringkali berasal dari sumber pencemaran seperti limbah pertanian, tampaknya tidak secara langsung mempengaruhi pH.

6. pH vs Chloride:

Hubungan antara klorida dan pH juga sangat lemah. Klorida yang umum ditemukan dalam air tidak banyak mempengaruhi keasaman atau kebasaan air.

7. pH vs Zinc:

Korelasi yang lemah antara zink dan pH menunjukkan bahwa zink, meskipun dapat terlarut dalam air, tidak cukup signifikan untuk mengubah pH secara drastis.

8. pH vs Turbidity:

Kekeruhan atau turbiditas menunjukkan korelasi negatif yang lemah dengan pH, yang bisa jadi menunjukkan bahwa air yang lebih keruh cenderung sedikit lebih asam, mungkin karena partikel yang tersuspensi.

9. pH vs Copper:

Tidak ada hubungan yang signifikan antara tembaga dan pH, menunjukkan bahwa konsentrasi tembaga tidak banyak mempengaruhi tingkat pH.

10. pH vs Iron:

Korelasi yang sangat lemah antara besi dan pH menunjukkan bahwa meskipun besi dapat mempengaruhi warna dan rasa air, tidak tampak mempengaruhi pH air secara signifikan.

11. pH vs Chlorine:

Meskipun klorin sering digunakan untuk desinfeksi, tidak terlihat adanya korelasi yang kuat dengan pH, berarti tingkat klorinasi air tidak secara langsung berkaitan dengan pH air.

12. pH vs Sulfate:

Sulfat menunjukkan korelasi negatif yang lemah dengan pH. Hal ini mungkin menunjukkan bahwa air dengan kandungan sulfat yang lebih banyak cenderung sedikit lebih asam, meskipun hubungan ini tidak kuat.

13. pH vs Manganese:

Tidak ada hubungan yang kuat antara mangan dan pH, yang berarti bahwa konsentrasi mangan dalam air tidak berpengaruh banyak terhadap pH.

14. pH vs Air Temperature:

Sama seperti suhu air, suhu udara juga tidak menunjukkan pengaruh yang signifikan terhadap pH air.
